```python
import numpy as np
import random

def crossover(arr1,arr2):
    return arr1[0:int(queens/2)]+arr2[int(queens/2):int(queens)]

def mutate(arr):
    # select target
    i = random.randint(0,queens-1)
    newVal = random.randint(0,queens-1)
    arr[i] = newVal
    return arr

def hasClash(index,arr):
    col_no = index
    row_no = arr[index]
    
    clashInRow = False
    clashInDiagonal = False
    
    #check in row
    for i in range(len(arr)):
        if (i != col_no and arr[i] == row_no):
            clashInRow = True
            break    
    
    #check in diagonal
    if not clashInRow:
        
        i=row_no+1
        j=col_no+1
        while i < len(arr) and j < len(arr):
            if(arr[j] == i):
                clashInDiagonal = True
                break
            i+=1
            j+=1
    
        if not clashInDiagonal:
            i=row_no-1
            j=col_no-1
            while (i > -1 and j > -1):
                if(arr[j] == i):
                    clashInDiagonal = True
                    break
                i-=1
                j-=1

        if not clashInDiagonal:
            i=row_no+1
            j=col_no-1
            while (i < len(arr) and j > -1):
                if(arr[j] == i):
                    clashInDiagonal = True
                    break
                i+=1
                j-=1

        if not clashInDiagonal:
            i=row_no-1
            j=col_no+1
            while (i > -1 and j < len(arr)):
                if(arr[j] == i):
                    clashInDiagonal = True
                    break
                i-=1
                j+=1


    return (clashInRow or clashInDiagonal)
    
def fitness(arr):
    noClashes = 0
    for i in range(len(arr)):
        if not hasClash(i,arr):
            noClashes += 1
    return noClashes

def printSolution(arr):
    print("Solution: ",arr)
    for i in range(0,queens):
        print("|",end="")
        for j in range(0,queens):
            if arr[j]==i:
                print("Q|",end='')
            else:
                print(" |",end='')
        print("")

```

```python
queens = 8
repeatCount = 100
populationSize = 50   # at least 6
keepCount = int(populationSize/2)      # at least half of populationSize

population = [random.choices(range(0,queens), k=queens) for _ in range(populationSize)]
# population[9] = [2,7,0,4,6,1,5,2]

count = 0

while(count<repeatCount):
    count+=1
    print("Iteration No. ",count)
#     print(np.matrix(population))
    
    population.sort(key=fitness,reverse=True)
    if fitness(population[0]) == queens:
        printSolution(population[0])
        break
    
    population = population[:keepCount]
    print("\tMax Goodness achieved: ",fitness(population[0]) ," out of ",queens) 
    print("\tfor population: ",population[0])

    
    i=keepCount
    while i<populationSize:
        temp = keepCount-1
        idx1 = random.randint(0,temp)
        idx2=idx1
        while idx2==idx1:
            idx2 = random.randint(0,temp)
        population.append(mutate(crossover(population[idx1],population[idx2])))
        i+=1
    
```
